```python
industry_summary = df.groupby('Industry').agg({
    'Revenue ($)':'sum',
    'Employees':'sum'
}).sort_values(by='Revenue ($)',ascending=False)
print('\n Top Industries by Revenue and Employees')
print(industry_summary.head(10))
```

```python
state_summary = df.groupby('State').agg({
    'Revenue ($)':'sum',
    'Employees':'sum'
}).sort_values(by='Revenue ($)',ascending=False)
print('\nTop States by Revenue and Employees')
print(state_summary.head(10))
```

```python
# Pivot table
pivot_table_industry_state = pd.pivot_table(df,values=['Revenue ($)','Employees'],index='Industry',columns='State',aggfunc='mean')
print('Average Revenue and Employees by Industry and State:')
print(pivot_table_industry_state.head(10))
```

```python
pivot_table_state_industry = pd.pivot_table(df,values=['Revenue ($)','Employees'],index='State',columns='Industry',aggfunc='mean')
print('Average Revenue and Employees by State and Industry')
print(pivot_table_state_industry.head())
```

```python
crosstab_industry_state = pd.crosstab(index=df['Industry'],columns=df['State'])
print('Crosstab of Industry by State (Number of companies):')
print(crosstab_industry_state.head())
```

```python
crosstab_industry_ceo = pd.crosstab(index=df['Industry'],columns=df['CEO'])
print('\nCrosstab of Industry by CEO(Number of Companies):')
print(crosstab_industry_ceo.head(1))
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Employees',y='Revenue ($)',data=df,hue='Industry',size='Revenue ($)',sizes=(20,1000),alpha=0.7)
plt.title('Revenue Vs Employees by Industry')
plt.xlabel('Number of Employees')
plt.ylabel('Revenue ($)')
plt.xscale('log')
plt.yscale('log')
plt.grid(True,which='both',ls='--',c='0.7')
plt.legend(bbox_to_anchor=(1.05,1),loc='upper left',borderaxespad=0.,fontsize='small')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='Rank',y='Revenue ($)',data=df,hue='Industry',size='Employees',sizes=(20,1000),alpha=0.7)
plt.title('Rank Vs Revenue by Industry (Point size by Employees)')
plt.xlabel('Company Rank')
plt.ylabel('Revenue ($)')
plt.yscale('log')
plt.grid(True,which='both',ls='--',c='0.7')
plt.legend(bbox_to_anchor=(1.05,1),loc='upper left',borderaxespad=0,fontsize='small')
plt.tight_layout()
plt.show()
```

```python
top_5_industries = industry_summary.head(5).index
filtered_data = df[df['Industry'].isin(top_5_industries)]

plt.figure(figsize=(15,10))
g = sns.FacetGrid(filtered_data,col='Industry',col_wrap=3,sharex=False,sharey=False)
g.map(sns.histplot,'Revenue ($)',bins=20,kde=True)
g.set_titles('Industry:{col_name}')
g.set_axis_labels('Revenue ($)')


for ax in g.axes.flat:
    ax.set_xscale('log')

plt.suptitle('Revenue Distribution Across Top 5 Industries (Histograms)', y=1.02) # Adjust suptitle position
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=industry_summary.head(10).index,y=industry_summary['Revenue ($)'].head(10),palette='viridis')
plt.title('Top 10 Industries by Revenue ($)')
plt.xlabel('Industry')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=state_summary.head(10).index,y=state_summary['Revenue ($)'].head(10),palette='magma')
plt.title('Top 10 states by revenue')
plt.xlabel('State')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
